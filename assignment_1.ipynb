# **Parsing Date **

```python
from datetime import datetime
from datetime import date
```

```python
airbnb_new_df.dtypes
```

**Dates conversion**

```python
airbnb_new_df['year']=pd.to_datetime(airbnb_new_df['last_review'] ,errors='coerce')
```

```python
airbnb_new_df['month'] = airbnb_new_df['year'].dt.strftime('%m')
```

**Handling missing data**

```python
airbnb_new_df.loc[pd.isnull(airbnb_new_df['year']), 'year'] = 0
```

```python
airbnb_new_df.loc[pd.isnull(airbnb_new_df['month']), 'month'] = 0
```

```python
airbnb_new_df
```

**Estimated Property Revenue**

```python
airbnb_new_df['estimated_revenue'] = airbnb_new_df['price'] * airbnb_new_df['minimum_nights'] * airbnb_new_df['calculated_host_listings_count']
```

```python
airbnb_new_df
```

```python
airbnb_new_df['estimated_revenue'].count()
```

```python
total_revenue_df =airbnb_new_df[['name','estimated_revenue']].groupby(['name'])
```

```python
revenue=total_revenue_df.head(10)
```

```python
revenue
```

```python
revenue.head(10).plot(kind='scatter', x='estimated_revenue',y='name')
```

**What are the busiest months for renters?**

```python
bookings_by_month = airbnb_new_df['year'].groupby(airbnb_new_df['month']).count()
bookings_by_month.plot.bar()
```
